```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os
```

```python
def wait_a_sec(driver):
    WebDriverWait(driver, 30).until(
    lambda driver: driver.execute_script('return document.readyState') == 'complete')

# Path to your browser driver (adjust this according to your setup)
driver_path = rf"C:\Users\simof\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe"

driver = webdriver.Chrome(service = Service(driver_path))

# Open the URL (replace with your target URL)
url = 'https://servizi.regione.piemonte.it/catalogo/banca-dati-agrometeorologica-ram'  # Replace with the actual URL of the page you're scraping
driver.get(url)

# Wait for the page to load and elements to become interactable
wait_a_sec(driver)

button = driver.find_element(By.XPATH, "//*[@target='_blank']")  # Find by button text
button.click()


windows = driver.window_handles
# Step 5: Switch to the new tab (the second tab)
driver.switch_to.window(windows[1])

button2 = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//*[text()='Consultazione e scaricamento dei dati di base']"))
)
button2.click()
wait_a_sec(driver)

button3 = driver.find_element(By.XPATH, "//*[text()='Consulta i dati giornalieri su un intervallo di tempo personalizzato']")
button3.click()
wait_a_sec(driver)
```

```python
def download_weather(driver, provincia, centralina, anno):
    select_element = driver.find_element(By.ID, "tipologia")
    select = Select(select_element)
    select.select_by_value("ELETTRONICA") 
    wait_a_sec(driver)
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.ID, "Provincia"))
    )

    select_element2 = driver.find_element(By.ID, "Provincia")
    select2 = Select(select_element2)
    select2.select_by_value(provincia)
    wait_a_sec(driver)
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, '//*[@value="{}"]'.format(centralina[-1])))
    )

    # s = driver.find_element(By.XPATH, '//*[@value="{}"]'.format(c))
    # s.click()

    select_element = driver.find_element(By.ID, "idCentralina")
    driver.implicitly_wait(10)
    select = Select(select_element)

        
    for o in select.options:
        v = o.get_attribute('value')
        if v in centralina:
            select.select_by_value(v)
    wait_a_sec(driver)

    button4 = driver.find_element(By.NAME, "bu_conferma_new")
    button4.click()

    wait_a_sec(driver)

    select_element = driver.find_element(By.ID, "Parametri")
    select = Select(select_element)

    for o in select.options:
        v = o.get_attribute('value')
        if not v.startswith('$'):
            select.select_by_value(v)

    button5 = driver.find_element(By.NAME, "bu_conferma")
    button5.click()
    wait_a_sec(driver)

    text_field = driver.find_element(By.ID, 'giorno')  # Replace with actual ID or XPath
    text_field.send_keys('01')  # Enter your data here
    text_field = driver.find_element(By.ID, 'mese')  # Replace with actual ID or XPath
    text_field.send_keys('Gennaio')  # Enter your data here
    text_field = driver.find_element(By.ID, 'anno')  # Replace with actual ID or XPath
    text_field.send_keys(anno)  # Enter your data here

    text_field = driver.find_element(By.ID, 'giornoF')  # Replace with actual ID or XPath
    text_field.send_keys('31')  # Enter your data here
    text_field = driver.find_element(By.ID, 'meseF')  # Replace with actual ID or XPath
    text_field.send_keys('Dicembre')  # Enter your data here
    text_field = driver.find_element(By.ID, 'annoF')  # Replace with actual ID or XPath
    text_field.send_keys(anno)  # Enter your data here

    button6 = driver.find_element(By.NAME, "bu_conferma_cen")
    button6.click()
    wait_a_sec(driver)

    button7 = driver.find_element(By.NAME, "bu_csv")
    button7.click()

    print(f"File {provincia},{centralina}, {anno} downloaded successfully")
```
