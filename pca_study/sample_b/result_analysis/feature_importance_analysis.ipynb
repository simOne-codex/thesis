```python
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('/nfs/home/genovese/thesis-wildfire-genovese/pca_study/database/sample_b/trained_models/final_regression.pkl', 'rb') as f:
    r_model = pickle.load(f)

with open('/nfs/home/genovese/thesis-wildfire-genovese/pca_study/database/sample_b/trained_models/final_classification.pkl', 'rb') as f:
    c_model = pickle.load(f)

with open ('/nfs/home/genovese/thesis-wildfire-genovese/pca_study/database/sample_b/trained_models/pca.pkl', 'rb') as f:
    pca = pickle.load(f)
    
X_train = pd.read_csv('/nfs/home/genovese/thesis-wildfire-genovese/pca_study/database/sample_b/final_X_train_val.csv')
old_features = pd.read_csv('/nfs/home/genovese/thesis-wildfire-genovese/pca_study/database/sample_b/X_train_val.csv').columns
```

```python
rf = pd.Series(r_model.feature_importances_, index=X_train.columns).sort_values(ascending=False)
cf = pd.Series(c_model.feature_importances_, index=X_train.columns).sort_values(ascending=False)
```

```python
pd.concat([pd.Series(rf.index, name='regression'), pd.Series(cf.index, name='classification')], axis=1).to_latex()
```

```python
print(rf, '\n \n')
print(cf)
```

```python
pca_transform = pd.DataFrame(pca.components_, columns = [f'PC{n}' for n in range(1, len(old_features)+1)], index=old_features)
r_loadings = pca_transform.loc[:, [pc for pc in rf.index if 'PC' in pc]]
c_loadings = pca_transform.loc[:, [pc for pc in cf.index if 'PC' in pc]]
```

```python
for col in pd.Series(list(r_loadings.columns)+list(c_loadings.columns)).unique():
    print(f"\nTop features for {col}:")
    top_features = r_loadings.loc[:, col].abs().sort_values(ascending=False).head(30)
    print(top_features)

for col in r_loadings.columns:
    print(f"\nTop features for {col}:")
    top_features = r_loadings.loc[:, col].abs().sort_values(ascending=False).head(30)
    print(top_features)
```
